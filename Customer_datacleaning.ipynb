```python
df['Review Rating'] = (
    df.groupby('Category')['Review Rating']
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df.columns
```

```python
bins = [0, 25, 40, 60, 100]
labels = ['Young Adult', 'Adult', 'Middle Age', 'Senior']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df[['age', 'age_group']].head()
```

```python
frequency_map = {
    'Daily': 1,
    'Weekly': 7,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_map)
df[['frequency_of_purchases', 'purchase_frequency_days']].head()
```

```python
df.drop(columns=['promo_code_used'], inplace=True)
```
